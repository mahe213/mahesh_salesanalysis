```python
import matplotlib.pyplot as plt
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.figure(figsize = (5,3),dpi = 200)
#plt.title('Graph',fontdict = {'fontname' : 'comic sans ms','fontsize': 20, 'color': 'red'})
plt.plot(x,y, label = '2x',color = 'yellow', linewidth = 2,marker = '.', markersize = 15, linestyle = '--', markeredgecolor = 'blue')
plt.plot(x,y,'b--',label = 'x2',color = 'yellow')
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,color = 'red', label = 'x2')
plt.plot(x2[5:],x2[5:]**2, color = 'red')
plt.xlabel('x axis')
plt.xlabel('Y axis')
plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show
```
